# Stock price and return time series

### Get data

```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats
import pydy
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('T1_ts_data_-1337276833.xlsx')
```

```python
data.set_index('Date',inplace = True)
```

```python
data
```

### Simple plots for FTSE100, BP, and Shell

```python
plt.plot(data['FTSE100'])
```

```python
plt.plot(data['BP'])
```

```python
plt.plot(data['SHELL'])
```

```python
plt.figure(figsize=(12,6))
```

```python
# Loop through each stock column (assuming 'BP', 'SHELL', and another)
for col in ['FTSE100','BP', 'SHELL']:
  plt.plot(data.index, data[col], label=col)  # Plot with label for legend

# Add legend to distinguish lines
plt.legend()

plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

bp, = plt.plot(data['BP'],label = 'BP',color = 'b')
plt.legend()
ax2 = plt.twinx()
shell, = ax2.plot(data['SHELL'],label = 'SHELL',color = 'g')
plt.legend(handles=[bp,shell])



plt.title('BP and Shell Stock Prices')
plt.xlabel('Date')
plt.ylabel('Prices')

```

### Plot the returns for each asset

```python
returns = data.pct_change()
```

```python
returns
```

```python
returns = returns[1:] #get rid of initial NAN
```

```python
plt.plot(returns['FTSE100'])
```

```python
plt.plot(returns['BP'])
```

```python
plt.plot(returns['SHELL'])
```

```python
plt.figure(figsize=(12,6))
```

```python
# Loop through each stock column (assuming 'BP', 'SHELL', and another)
for col in ['FTSE100','BP', 'SHELL']:
  plt.plot(returns.index, returns[col], label=col)  # Plot with label for legend

# Add legend to distinguish lines
plt.legend()

plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.grid(True)
plt.tight_layout()
plt.show()
```
